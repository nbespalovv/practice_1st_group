```python
import requests
from bs4 import BeautifulSoup
import sqlite3
import re
```

```python
class Crawler:
    def __init__(self, url):
        self.url = url
    
    def get_soup(self):
        soup = BeautifulSoup(requests.get(self.url).text, "html.parser")
        return soup
```

```python
class Parser:
    def get_info():

        for page in range(38):
            soup = Crawler(f'https://ru.kinorium.com/movies/online/?order=rating&page={page}&perpage=200&show_viewed=1').get_soup()

            for film in soup.findAll('div', class_ = 'info-box filmList__item-title-wrap'):
                name = film.find('i', class_ = 'movie-title__text').text.replace(u'\xa0', u' ').strip()
                link = 'https://ru.kinorium.com' + film.find('a', class_ = 'filmList__item-title').get('href')
                print(name)

                soup_film = Crawler(link).get_soup()
                description = soup_film.find('section', class_ = 'text film-page__text').text.replace(u'\xa0', u' ')

                countries = []
                for country in soup_film.findAll('a', class_ = 'film-page__country-link'):
                    countries.append(country.text)

                genres = []
                for genre in soup_film.findAll('li', itemprop = 'genre'):
                    genres.append(genre.text)

                soup_cast = Crawler(link + 'cast').get_soup()

                employees = {}
                for t in soup_cast.findAll('div', class_='ref-list clearfix'):
                    for pos in t.findAll('h1', class_='cast-page__title'):
                        position = pos.text.strip()
                        employees[position] = []
                        for emp in t.findAll('h5', class_='cast-page__item-name'):
                            employee = emp.text.strip()
                            employees[position].append(employee)
                
                #этот массив нужно добавлять в бд
                info = [name, link, description, countries, genres, employees]
                print(info)
```
